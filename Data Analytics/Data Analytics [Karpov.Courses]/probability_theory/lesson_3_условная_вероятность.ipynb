### Какова вероятность, что стрелок был из первой группы, если он попал в цель?

P(H1) – вероятность гипотезы №1, например, что стрелок из 1-ой группы попал в мишень

P(A) – формула полной вероятности, вероятность попадания в целом всех, вероятность изготовления в целом всех

По факту это n/m = вероятность гипотезы на полную вероятность события
