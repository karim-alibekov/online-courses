```python
df_install = df.query("event == 'app_install'")
df_install = df_install.rename(columns={'date':'install_date'})
df_install = df_install.drop(columns=['event', 'purchase_sum', 'os_name', 'gender', 'city', 'utm_source', 'month'])
df_install
```

```python
df_purchase = df.query("event == 'purchase'")
df_purchase = df_purchase.rename(columns={'date':'purchase_date'})
df_purchase = df_purchase.drop(columns=['event', 'purchase_sum', 'os_name', 'gender', 'city', 'utm_source', 'month'])
df_purchase
```

```python
df_conversion = pd.merge(df_install, df_purchase, on='device_id')
df_conversion
```

```python
df_conversion = df_conversion.groupby(['device_id', 'install_date'], as_index=False).agg({'purchase_date':'min'})
df_conversion
```

```python
df_conversion['time_diff'] = df_conversion.purchase_date - df_conversion.install_date
df_conversion
```

```python
df_conversion_2 = df_conversion.query("time_diff <= '7 days'").groupby('install_date', as_index = False).agg({'purchase_date':'count'})
df_conversion_2 = df_conversion_2.rename(columns={'install_date':'date', 'purchase_date':'purchases_within_7_days_from_install'})
df_conversion_2
```

```python
df_all_installs_by_day = df.query('event == "app_install"').groupby('date', as_index = False).agg({'device_id':'count'})
df_all_installs_by_day = df_all_installs_by_day.rename(columns={'device_id':'installs'})
df_all_installs_by_day
```

```python
final_df = pd.merge(df_all_installs_by_day, df_conversion_2, on='date')
final_df['CR'] = final_df.purchases_within_7_days_from_install / final_df.installs
final_df.sort_values(by='CR', ascending=False)
```
