```python
money_by_city = df.groupby(['title', 'city'], as_index=False) \
    .aggregate({'money': 'sum'}) \
    .sort_values('money', ascending=False)
```

## Выше – подсчет всех заказов, ниже – подсчет заказов со статусом "завершен"

```python
money_by_title = df \
    .query("status == 'Завершен'") \
    .groupby(['title'], as_index=False) \
    .aggregate({'money': 'sum', 'number': 'count'}) \
    .sort_values('money', ascending=False) \
    .rename(columns={'number': 'success_orders'})
```

```python
money_by_title
```

## Самопроверка после анализа

```python
if int(money_by_title.money.sum()) == int(all_money):
    print('OK')
else:
    print('ERROR!!!')
```

## Автоматизация скучной работы

```python
today_day = datetime.today().strftime('%Y-%m-%d')
file_name = 'money_by_title_{}.csv'
file_name = file_name.format(today_day)
```

```python
if int(money_by_title.money.sum()) == int(all_money):
    print('OK! File {} is written.'.format(file_name))
    money_by_title.to_csv(file_name, index=False)
else:
    print('ERROR!!!')
```

### Аргумент file_name в money_by_title.to_csv(**file_name**, index=False) : это literally имя файла
