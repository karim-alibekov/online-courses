```python
path_to_file = '/mnt/HC_Volume_18315164/home-jupyter/jupyter-k-alibekov/shared/Res_Tree/F000545/res_2019.09.11_0.0_6493B6_Container-dat_2125_91-105-165_F000545/meals_list_2.txt'
```

```python
all_data = []
```

```python
scanning = False

for i in open(path_to_file).readlines():
    
    if i.startswith('--------------'):
        scanning = False
        continue
    
    if scanning:
        if 'Meal#;' in i:
            col_name = i
            continue
        all_data.append([value.strip() for value in i.split(';')])
    
    if 'Non-deleted saved meals' in i:
        scanning = True
        continue


# open(path_to_file).readlines()
```

```python
col_name
```

```python
final_col_names = [i.strip() for i in col_name.split(';')]
```

```python
final_col_names
```

```python
all_data #список из списков
```

```python
meal_df = pd.DataFrame(all_data, columns=final_col_names[0:17])
```

```python
meal_df
```

```python
meal_df.dtypes
```

```python
meal_df = meal_df.astype({'Minutes': 'int', 'Pro': 'float'})
```

```python
meal_df['Date'] = pd.to_datetime(meal_df.Date, dayfirst=True)
```

```python
user_meal_df = meal_df[['Date', 'Pro', 'Minutes']]
```

```python
user_meal_df
```

```python
user_meal_df.dtypes
```

```python
user_meal_df.Date.dt.daysinmonth
```

```python
df = user_meal_df.groupby('Date', as_index=False).agg({'Minutes': 'sum'})
```

```python
df.to_csv('df_minutes.csv', index=False)
```
