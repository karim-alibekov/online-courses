```python
new_df.ACT_DTTM = pd.to_datetime(new_df['ACT_DTTM'], format='%d-%m-%Y %H:%M')
new_df.START_DTTM = pd.to_datetime(new_df['START_DTTM'], format='%d-%m-%Y %H:%M')
new_df.END_DTTM = pd.to_datetime(new_df['END_DTTM'], format='%d-%m-%Y %H:%M')
```

### ВНИМАНИЕ! format задаёт только то, как считываются данные. Отображаться в самом датафрейме дата всегда будет в американском формате год-месяц-день. Поэтому не пугайтесь того, что вы написали ему день-месяц-год, а пишет он в другом порядке - это нормально.

```python
new_df
```

```python
new_df['time_difference'] = new_df.END_DTTM - new_df.START_DTTM
```
