## делаем try, except потому что для некоторых объектов может не быть информации + для конката в конце квадратные скобочки

### Потому что не все из этих случаев будут работать. Есть некоторые объекты, которые не содержат часть нужных нам полей - соответственно, при попытке конвертации в датафрейм с такими случаями могут возникать ошибки. Именно их и обрабатывает конструкция try-except.

```python
for i in sunflower_ids:
    
    try:
        
        a = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}')
        x = a.json()
        df_cycle = pd.DataFrame({'objectID':[x['objectID']], 
                       'title':[x['title']], 
                       'artistDisplayName':[x['artistDisplayName']], 
                       'department':[x['department']],
                       'objectBeginDate':[x['objectBeginDate']], 
                       'objectEndDate':[x['objectEndDate']], 
                       'period':[x['period']], 
                       'objectName':[x['objectName']], 
                       'culture':[x['culture']]})
        final_df = pd.concat([final_df, df_cycle])
    except:
        continue
```

```python
final_df.groupby('culture', as_index=False).agg({'objectID':'count'})
```

```python
r = requests.get('https://collectionapi.metmuseum.org/public/collection/v1/search?departmentId=6&q=cat')
cats = r.json()
```

```python
cats_ids = cats['objectIDs']
cats_ids
```

```python
cats_df = pd.DataFrame(columns=['objectID', 
                                 'title', 
                                 'artistDisplayName', 
                                 'department', 
                                 'objectBeginDate', 
                                 'objectEndDate', 
                                 'period', 
                                 'objectName', 
                                 'culture'])
```

```python
cats_df
```

```python
for i in cats_ids[0:1000]:
    
    try:
        
        a = requests.get(f'https://collectionapi.metmuseum.org/public/collection/v1/objects/{i}')
        x = a.json()
        df_new = pd.DataFrame({'objectID':[x['objectID']], 
                       'title':[x['title']], 
                       'artistDisplayName':[x['artistDisplayName']], 
                       'department':[x['department']],
                       'objectBeginDate':[x['objectBeginDate']], 
                       'objectEndDate':[x['objectEndDate']], 
                       'period':[x['period']], 
                       'objectName':[x['objectName']], 
                       'culture':[x['culture']]})
        cats_df = pd.concat([cats_df, df_new])
    except:
        continue
```

```python
cats_df.groupby('period').agg({'objectID':'count'}).sort_values('objectID', ascending=False)
```
