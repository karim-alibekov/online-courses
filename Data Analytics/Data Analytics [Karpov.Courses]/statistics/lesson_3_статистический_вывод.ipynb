```python
import pandas as pd
import numpy as np

import seaborn as sns
%matplotlib inline
```

```python
mu = 100 #mean
sigma = 10 #standard deviation
```

```python
population = np.round(np.random.normal(mu, sigma, 10000), 2)
```

```python
sns.displot(population, kde=False)
```

```python
population
```

```python
population.mean()
```

```python
population.std()
```

```python
sample_size = 30
```

```python
np.random.choice(population, sample_size, False)
```

```python
sample = np.random.choice(population, sample_size, False)
```

```python
sample.mean()
```

```python
sample_means = []
```

```python
n = 10000
```

```python
for i in range(n):
    sample = np.random.choice(population, sample_size, False)
    sample_means.append(sample.mean())
```

```python
# РАСПРЕДЕЛЕНИЕ СРЕДНИХ ЗНАЧЕНИЙ В СИМУЛЯЦИОННЫХ ВЫБОРКАХ // 10.000 ВЫБОРОК ПО 30 НАБЛЮДЕНИЙ

sns.displot(sample_means)
```

```python
np.mean(sample_means)
```

```python
# правило 3 сигм – 100% наблюдений в этом диапазоне
# правило 1.96 сигм – 95% наблюдений в этом диапазоне
```

```python
# se = sigma / root of n 
```

```python
se = population.std() / sample_size ** 0.5
se
```

```python
np.std(sample_means)
```

```python
population.mean() - 1.96 * se
```

```python
population.mean() + 1.96 * se
```
